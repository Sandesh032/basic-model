```python
import pandas as pd

# Extract csv content
nq_df = pd.read_csv("nq.csv")
gc_df = pd.read_csv("gc.csv")
```

```python
# Format the date and time

nq_df['Time'] = pd.to_datetime(nq_df['Time'].astype(str).str.zfill(6), format='%H%M%S').dt.time
nq_df['Date'] = pd.to_datetime(nq_df['Date'].astype(str), format='%Y%m%d').dt.date

gc_df['Time'] = pd.to_datetime(gc_df['Time'].astype(str).str.zfill(6), format='%H%M%S').dt.time
gc_df['Date'] = pd.to_datetime(gc_df['Date'].astype(str), format='%Y%m%d').dt.date
```

```python
# Combine date-time as single column

nq_df['Datetime'] = pd.to_datetime(nq_df['Date'].astype(str) + ' ' + nq_df['Time'].astype(str))
nq_df = nq_df[['Datetime', 'Price']].set_index('Datetime').sort_index()
```

```python
gc_df['Datetime'] = pd.to_datetime(gc_df['Date'].astype(str) + ' ' + gc_df['Time'].astype(str))
gc_df = gc_df[['Datetime', 'Price']].set_index('Datetime').sort_index()
```
